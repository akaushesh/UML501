```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("Import done")

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
df = pd.DataFrame(columns=['Title', 'Price', 'Availability', 'Star Rating'])

rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

page = 1
while True:
    if page == 1:
        url = "https://books.toscrape.com/"
    else:
        url = base_url.format(page)
    
    response = requests.get(url)
    if response.status_code != 200:
        break

    soup = BeautifulSoup(response.text, 'html.parser')

    titles = [tag.a['title'] for tag in soup.find_all('h3')]
    prices = [tag.text.strip() for tag in soup.find_all('p', class_='price_color')]
    availabilities = [tag.text.strip() for tag in soup.find_all('p', class_='instock availability')]
    ratings = [rating_map[tag['class'][1]] for tag in soup.select('p.star-rating')]

    if not titles:
        break

    temp_df = pd.DataFrame({
        'Title': titles,
        'Price': prices,
        'Availability': availabilities,
        'Star Rating': ratings
    })

    df = pd.concat([df, temp_df], ignore_index=True)
    print("Page done: ",page)
    page += 1

df.to_csv('books.csv', index=False)
```
